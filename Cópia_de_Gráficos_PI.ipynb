```python
import pandas as pd
import seaborn as se
```

```python
agendamento = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/agendamento3.csv")
```

```python
agendamento.head(100)
```

```python
import matplotlib.pyplot as plt

agendamento['Curso'].value_counts().plot(kind='bar', color=['blue','red','green'])
plt.title('Agendamentos por Curso')
plt.xlabel('Curso')
plt.ylabel('Quantidade')
plt.show()
```

```python
agendamento['Módulo'].value_counts().sort_index().plot(kind='bar', color=['orange','green','red'])
plt.title('Agendamentos por Módulo')
plt.xlabel('Módulo')
plt.ylabel('Quantidade')
plt.show()
```

```python
agendamento['Mês'].value_counts().sort_index().plot(kind='bar',
color=['orange','green','red', 'yellow', 'Blue', 'black'])
plt.title('Agendamentos por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.show()
```
